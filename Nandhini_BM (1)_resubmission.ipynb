```python
import matplotlib.pyplot as plt

# Model stages
stages = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']
metrics = {
    'Dummy':                  (0.4767, 0.5075, 0.5048, 0.5327),
    'KNN Training':           (0.8983, 0.8950, 0.8867, 0.9045),
    'KNN Prediction':         (0.9717, 0.963, 0.9407, 0.9061),
    'KNN Retraining':         (0.8983, 0.721, 0.4163, 1.0),
    'KNN Retrained Prediction': (0.9717, 0.9650, 0.913, 0.8873)
}

train_accuracy = [metrics[stage][0] for stage in stages]
test_accuracy = [metrics[stage][1] for stage in stages]
precision = [metrics[stage][2] for stage in stages]
recall = [metrics[stage][3] for stage in stages]
```

```python
plt.figure(figsize=(12, 8))
plt.plot(stages, train_accuracy, 'o-', color='blue', label='Train Accuracy')
plt.plot(stages, test_accuracy, 'o-', color='green', label='Test Accuracy')
plt.plot(stages, precision, 'o-', color='red', label='Precision')
plt.plot(stages, recall, 'o-', color='purple', label='Recall')
plt.xlabel('Model Stages')
plt.ylabel('Performance Metric')
plt.title('Model Performance Across Different Training Stages')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.xticks(rotation=15)

for i, stage in enumerate(stages):
    plt.text(i, train_accuracy[i] + 0.01, f'{train_accuracy[i]:.2f}', ha='center')
    plt.text(i, test_accuracy[i] - 0.02, f'{test_accuracy[i]:.2f}', ha='center')
    plt.text(i, precision[i] + 0.01, f'{precision[i]:.2f}', ha='center')
    plt.text(i, recall[i] - 0.02, f'{recall[i]:.2f}', ha='center')
plt.ylim(0.4, 1.0)
plt.tight_layout()
plt.show()
```
